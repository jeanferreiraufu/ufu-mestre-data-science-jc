  Resposta:

    A digitalização de informações analógicas envolve a conversão dessas informações em formato digital, onde os dados são representados em números binários (0s e 1s). Os passos gerais são:

    Captura: A informação analógica (som ou imagem) é capturada usando dispositivos como microfones para áudio ou câmeras para imagens.

    Codificação: Os valores quantizados são convertidos em um formato binário para armazenamento ou processamento digital.

## Áudio

### 1. O que é o som?
  Resposta:

    Som é uma onda mecânica que resulta da vibração de partículas em um meio (como ar, água ou metal). Essas vibrações criam compressões e rarefações no meio, que se propagam como uma onda sonora. O som é percebido quando essas ondas alcançam nossos ouvidos e são interpretadas pelo cérebro.

### 2. O que é amostragem e quantização de um sinal?
  Resposta:

    Amostragem é o processo de capturar valores de um sinal analógico contínuo em intervalos regulares de tempo, convertendo-o em um conjunto de valores discretos.
    Quantização é o processo de atribuir a cada amostra capturada um valor numérico, dentro de um intervalo limitado, que representa o sinal original. A quantização converte os valores contínuos em valores discretos.

### 3. O que é taxa de amostragem?
  Resposta:
  
    A taxa de amostragem é a frequência com que as amostras de um sinal são capturadas por segundo. Ela é medida em Hertz (Hz). Por exemplo, uma taxa de amostragem de 44,1 kHz significa que 44.100 amostras do sinal são capturadas a cada segundo.

### 4. O que é o teorema de Nyquist? Por que a taxa de Nyquist é importante durante o processo de digitalização de um sinal?
  Resposta:

    O Teorema de Nyquist afirma que para representar completamente um sinal analógico sem perda de informação ao digitalizá-lo, a taxa de amostragem deve ser pelo menos o dobro da maior frequência presente no sinal. A taxa de Nyquist é crucial porque, se a taxa de amostragem for inferior a ela, ocorre aliasing, onde as frequências do sinal são incorretamente representadas, causando distorção.

### 5. O que é aliasing?
  Resposta:

    Aliasing é um efeito indesejado que ocorre quando um sinal é amostrado a uma taxa inferior à taxa de Nyquist, fazendo com que diferentes sinais se tornem indistinguíveis uns dos outros após a amostragem. Isso resulta em uma distorção do sinal original.

### 6. Como o Teorema de Nyquist se aplica ao áudio e ao vídeo digitais? O que acontece se você tiver uma amostragem acima da taxa de Nyquist?
  Resposta:

    - Aplicação ao Áudio e Vídeo: No áudio digital, o teorema de Nyquist é aplicado para garantir que todas as frequências do som sejam corretamente representadas. Por exemplo, para um áudio com frequências até 20 kHz, a taxa de amostragem deve ser pelo menos 40 kHz. No vídeo, o conceito é semelhante, garantindo que a amostragem capture todos os detalhes visuais sem distorções.
    
    - Amostragem acima da taxa de Nyquist: Se a amostragem estiver acima da taxa de Nyquist, o sinal será representado com precisão, mas sem necessidade de melhorar a qualidade perceptível. No entanto, taxas muito altas podem resultar em arquivos maiores sem ganho perceptível na qualidade.

### 7. O que é modulação por código de pulso (PCM)? Como o PCM funciona?
  Resposta:

    PCM (Pulse Code Modulation) é um método para digitalizar sinais analógicos. Ele funciona através da amostragem do sinal em intervalos regulares e, em seguida, quantificando as amostras em valores discretos, que são representados em formato binário. Este método é amplamente utilizado em CDs de áudio e outros sistemas digitais.

### 8. Quais são os diferentes tipos de PCM? Descreva o funcionamento sobre os modelos: Differential PCM (DPCM) e Adaptive DPCM (ADPCM).
  Resposta:

    - DPCM (Differential PCM): Em vez de codificar o valor absoluto de cada amostra, o DPCM codifica a diferença entre a amostra atual e a anterior. Isso pode reduzir a quantidade de dados necessária.
    
    - ADPCM (Adaptive DPCM): É uma variação do DPCM que ajusta dinamicamente a escala de quantização com base nas características do sinal, permitindo uma compressão mais eficiente.

### 1. Mostre como a equalização de histograma pode ser aplicada em uma imagem colorida?
  Resposta:

    A equalização de histograma é uma técnica usada para melhorar o contraste de uma imagem. Para imagens coloridas, que são compostas de três canais (Red, Green, Blue), a equalização de histograma pode ser aplicada de duas formas:

      - Equalização por Canal: A equalização de histograma é aplicada separadamente a cada canal de cor (R, G, B). Isso pode resultar em mudanças nas cores da imagem.
      - Equalização no Canal de Intensidade: A imagem é convertida para o espaço de cores YUV ou HSV, onde o canal de intensidade (Y em YUV ou V em HSV) é equalizado. Os outros canais que contêm a informação de cor (U, V ou H, S) permanecem inalterados.

### 2. Qual a diferença entre resolução espacial e profundidade de imagens?
  Resposta:

    - Resolução Espacial: Refere-se à quantidade de detalhes em uma imagem, ou seja, o número de pixels em uma determinada área. Quanto maior a resolução espacial, mais detalhes a imagem possui. Por exemplo, uma imagem de 1024x1024 pixels tem uma maior resolução espacial que uma de 512x512 pixels.
    - Profundidade de Imagem: Refere-se ao número de bits usados para representar a cor de cada pixel. Uma profundidade maior significa mais tons de cor ou níveis de cinza disponíveis. Por exemplo, uma imagem de 8 bits por pixel pode representar 256 cores diferentes, enquanto uma de 24 bits pode representar 16,7 milhões de cores.  

### 3. Diferencie as técnicas de limiarizacão global e local.
  Resposta:

    - Limiarização Global: Usa um único valor de limiar para toda a imagem para segmentar objetos do fundo. É mais simples, mas pode ser ineficaz em imagens com iluminação desigual.
    - Limiarização Local: Aplica diferentes valores de limiar em diferentes regiões da imagem, levando em consideração a variação local de intensidade. Isso melhora a segmentação em imagens com variações de iluminação.

### 4. Diferencie modelos de cores subtrativos e aditivos.
  Resposta:

    - Modelos Aditivos (RGB): As cores são criadas pela adição de luzes de cor primária (Vermelho, Verde, Azul). É usado em dispositivos que emitem luz, como monitores e TVs.
    - Modelos Subtrativos (CMYK): As cores são criadas pela subtração de luz através da mistura de tintas ou pigmentos (Ciano, Magenta, Amarelo, e Preto). É usado em impressões.
